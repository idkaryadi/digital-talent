## Deskripsi Data

Untuk tugas ini dataset diambil dari data.go.id yang disajikan oleh Kementerian Kesehatan RI. Dataset ini berisi mengenai Indeks Pembangunan Manusia (IPM) di Indonesia dari tahun 2004 hingga 2012. Indeks Pembangunan Manusia (IPM) / Human Development Index (HDI) sendiri ialah suatu pengukuran perbandingan dari harapan hidup, melek huruf, pendidikan dan standar hidup untuk semua negara seluruh dunia. IPM digunakan untuk mengklasifikasikan apakah sebuah negara adalah negara maju, negara berkembang atau negara terbelakang dan juga untuk mengukur pengaruh dari kebijaksanaan ekonomi terhadap kualitas hidup.(Sumber: Wikipedia)

Pada Dataset ini digunakan variabel variabel sebagai berikut:

1. nama_provinsi: Nama Provinsi
2. kode_kabkota: Kode Kabupaten/Kota
3. nama_kabkota: Nama Kabupaten/Kota
4. tahun: Tahun
5. indikator_ipm: Indikator dari Indeks Pembangunan Manusia (IPM)
6. satuan: Satuan dari Indikator Indeks Pembangunan Manusia (IPM)
7. nilai: Nilai dari Indikator Indeks Pembangunan Manusia (IPM) tersebut

### Dataset

```python
import pandas as pd

#membaca dan menampilkan sampel data
data_ipm = pd.read_csv('ipm-2004-2012.csv', delimiter = ',', encoding = "ISO-8859-1")
data_ipm.sample(20)
```

```python
#mengecek nama kolom dataset
data_ipm.columns
```

```python
#mengelompokkan data nilai berdasarkan tahun, indikator_ipm serta satuannya
data_ipm.groupby(['tahun','indikator_ipm','satuan']).mean()['nilai']
```

```python
#mengecek entry dataset
data_ipm.count()
```

## Narasi
Apabila kita cermati, dataset ini memiliki nama kolom yang sesuai dengan variabel yang telah disebutkan diatas. Di kolom provinsi diisi oleh nama nama provinsi di Indonesia, sebab data ini bersifat nasional. Setelah kolom provinsi terdapat kolom kabupaten dan kota di provinsi tersebut. Ini menunjukkan bahwa data ditampilkan sampai ke tingkat kabupaten dan kota di Indonesia. Dengan demikian, dari data ini kita bisa mengamati perkembangan nilai ipm sampai ketingkat kabupaten/kota. Di kolom indikator ipm terdapat 4 indikator ipm beserta nilai ipm nya. Disebelahnya terdapat satuan dari indikator ipm dan diikuti oleh kolom nilai dari ipm berserta indikatornya. Sayangnya ada beberapa nilai tidak ada datanya atau kosong di tabel ini.

## Statistika Deskriptif

### Pivot Tabel Perkembangan Indikator IPM dari 2004-2011

```python
#membuat pivot tabel dengan indek tahun dan kolom indikator_ipm, berisi rata rata masing masing indikator setiap tahunnya
import numpy as np
table_tahun = pd.pivot_table(data_ipm, values='nilai', index=['tahun'],columns=['indikator_ipm'], aggfunc=np.mean)
table_tahun
```

### Pivot Tabel Perkembangan IPM tiap Provinsi dari 2004-2011

```python
#membuat pivot tabel dengan indek nama_provinsi dan kolom indikator_ipm, berisi rata rata masing masing indikator setiap tahunnya
df_table_prop = data_ipm.loc[(data_ipm['indikator_ipm']=='IPM')]
table_prop = pd.pivot_table(df_table_prop, values='nilai', index=['nama_provinsi'],columns=['tahun'], aggfunc=np.mean)
table_prop
```

### Pivot Tabel Statistik Deskriptif Indikator IPM

```python
#membuat pivot tabel statistik deskriptif untuk masing masing indikator_ipm akumulasi data 2004-2011
table_all = pd.pivot_table(data_ipm, values='nilai', index=['indikator_ipm'],columns=[], aggfunc={'nilai': [min, max, np.median, np.mean, np.std ]})
table_all
```

## Narasi
Apabila kita perhatikan, nilai IPM Indonesia mengalami peningkatan dari tahun ke tahun. Pada tahun 2004 nilainya sebesar 67.83, pada tahun 2011 nilainya menjadi sebesar 71.66. Nilai IPM naik karena semua indikator IPM mengalami kenaikan. Angka harapan hidup dari 67.11 tahun meningkat menjadi 68.90 tahun. Pengeluaran perkapita awalnya Rp 606.459 meningkat menjadi Rp 630.253. Rata rata lama sekolah awalnya 7.28 tahun meningkat menjadi 8.01 tahun.
Jika melihat dari 2004 hingga 2011, rata rata nilai IPM Indonesia ialah 69.94. Nilai maksimalnya 80.24, minimalnya 45.70 dengan standar deviasi 5.16. Nilai pencilan tersebut bisa dibilang nilai pencilan bawah. Ini memperlihatkan kepada kita bahwa masih ada daerah yang cukup tertinggal apabila dilihat dari nilai IPM nya. Daerah ini perlu dibangun agar tidak tertinggal terlalu jauh.
Dari data, yang cukup memprihatikan ialah rata rata lama sekolah yakni 7.65 tahun. Ini memperlihatkan kepada kita bahwa hampir sebagian besar penduduk di Indonesia tidak lulus SMP. Bahkan ada daerah yang rata rata lama sekolah 1.81, tidak sampai 2 tahun bersekolah formal. Bahkan ada daerah yang angka melek hurufnya cukup rendah, tidak sampai 30% penduduknya yang melek huruf. # Visualisasi Sederhana

```python
#membuat tabel nilai IPM berdasarkan tahun
df = data_ipm.loc[(data_ipm['indikator_ipm'] == 'IPM')]
table = pd.pivot_table(df, values='nilai', index=['tahun'],columns=['indikator_ipm'], aggfunc=np.mean)

#memvisualisasikan grafik nilai IPM berdasarkan tahun
line = table.plot(kind='line', title ="Perkembangan IPM berdasarkan Tahun", legend=True, fontsize=12, figsize=[20,10])
line.set_xlabel("Tahun", fontsize=12)
line.set_ylabel("Nilai", fontsize=12)
```

## Narasi
Dari grafik dapat diperhatikan bahwa nilai IPM Indonesia mengalami peningkatan dari tahun 2004 hingga 2011. Dari tampilannya sepertinya bisa dilakukan regresi linier untuk memprediksi nilai IPM di tahun selanjutnya. Hanya saja, saya belum bisa menampilkannya.
